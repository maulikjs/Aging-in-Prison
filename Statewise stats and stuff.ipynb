```python
yearend =pd.read_pickle('yearend.pkl')
tr=pd.read_pickle('tr.pkl')
releases=pd.read_pickle('releases.pkl')
admissions=pd.read_pickle('admissions.pkl')
```

```python
#get counts of total inmates by state
yearend = yearend[['SEX', 'RPTYEAR', 'STATE', 'EDUCATION', 'ADMTYPE', 'OFFGENERAL','ADMITYR','SENTLGTH', 'OFFDETAIL', 'RACE', 'AGEADMIT', 'AGEYREND','TIMESRVD_YREND']]
tr= tr[['ABT_INMATE_ID', 'SEX', 'ADMTYPE', 'OFFGENERAL', 'EDUCATION', 'ADMITYR','RELEASEYR','SENTLGTH', 'OFFDETAIL', 'RACE', 'AGEADMIT', 'AGERELEASE', 'TIMESRVD','RELTYPE', 'STATE']]
admissions = admissions[['SEX', 'RPTYEAR', 'STATE', 'EDUCATION', 'ADMTYPE', 'OFFGENERAL','ADMITYR','SENTLGTH', 'OFFDETAIL', 'RACE', 'AGEADMIT']]
releases = releases[['SEX', 'RPTYEAR', 'STATE', 'EDUCATION', 'ADMTYPE', 'OFFGENERAL','ADMITYR','SENTLGTH', 'OFFDETAIL', 'RACE', 'AGEADMIT', 'RELYR', 'RELTYPE','AGERLSE', 'TIMESRVD_REL']]

total_by_state = tr.groupby('STATE')['ABT_INMATE_ID'].count()
```

```python
tr_old_admit = tr[tr['AGEADMIT']==5]
```

```python
#Most common reason old people are going to prison in each state and what percentage they make up of the total no of old people going to prison
for each in states:
    newtr = tr_old_admit[tr_old_admit['STATE']==each]
    most_common_type=newtr['OFFDETAIL'].value_counts().head(1).index.tolist()
    if(len(most_common_type)!=0):
        print('state:',each)
        print(most_common_type[0], newtr['OFFDETAIL'].value_counts().head(1).values.tolist()[0]*100/newtr.shape[0])
        print('\n')
#12 drugs
#13 public order
#3  Rape/Sexual assault
```

```python
races=[1,2,3,4,9]
```

```python
#Most common reason old people are going to prison in each state and what percentage they make up of the total no of old people going to prison
print('Statewise by Race STATS')
for each in states:
    newtr = tr_old_admit[tr_old_admit['STATE']==each]
    print('<--------state:',each,'-------->')
    for eachrace in races:
        racetr = newtr[newtr['RACE']==eachrace]
        most_common_ethnicity=racetr['OFFDETAIL'].value_counts().head(1).index.tolist()
        if(len(most_common_ethnicity)!=0):
            print('RACE:',eachrace, '|| Percentage:',racetr.shape[0]*100/newtr.shape[0])
            print(most_common_ethnicity[0], racetr['OFFDETAIL'].value_counts().head(1).values.tolist()[0]*100/racetr.shape[0])
            print('\n')
        
#12 drugs
#13 public order
#3  Rape/Sexual assault
```

```python
#Most common reason old people are going to prison nationwided and what percentage they make up of the total no of old people going to prison
print('Nationwide By Race STATS')
for eachrace in races:
    racetr = tr_old_admit[tr_old_admit['RACE']==eachrace]
    most_common_ethnicity=racetr['OFFDETAIL'].value_counts().head(1).index.tolist()
    if(len(most_common_ethnicity)!=0):
        print('RACE:',eachrace, '|| Percentage:',racetr.shape[0]*100/tr_old_admit.shape[0])
        print(most_common_ethnicity[0], racetr['OFFDETAIL'].value_counts().head(1).values.tolist()[0]*100/racetr.shape[0])
        print('\n')
        
#12 drugs
#13 public order
#3  Rape/Sexual assault
```
